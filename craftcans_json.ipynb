```python
import requests
from BeautifulSoup import *
```

```python
url = ("http://craftcans.com/db.php?search=all&sort=beerid&ord=desc&view=text")
```

```python
response = requests.get(url)
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page)
```

```python
table = soup.findAll("table", attrs = {"style":"width:100%;margin-top:10px;"})
```

```python
print (table)
```

```python
data_list = []
all_rows = table[0].findAll('tr')
for row in all_rows:
    columns = row.findAll('td')
    beer = {
        "id":columns[0].text,
        "name":columns[1].find('b').text,
        "brewery":columns[2].text,
        "location":columns[3].text,
        "style":columns[4].text,
        "size":columns[5].text,
        "abv":columns[6].text,
        "ibu":columns[7].text
    }
    data_list.append(beer)
```
